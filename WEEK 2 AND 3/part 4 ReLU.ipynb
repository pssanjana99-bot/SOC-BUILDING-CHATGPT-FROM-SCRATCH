Why Activation Functions Are Needed

Without activation functions, every layer in a neural network would perform only linear transformations.

A network containing many linear layers is mathematically equivalent to a single linear transformation.

As a result:

The network cannot learn non-linear patterns.
Complex datasets cannot be modeled effectively.
The network behaves similarly to linear regression.

Activation functions introduce non-linearity and allow neural networks to learn complex relationships.

The Drug Dosage Dataset
A simple dataset involving drug dosages.

The observations show:

Low Dosage     → Ineffective
Medium Dosage  → Effective
High Dosage    → Ineffective

This produces a non-linear relationship.

A straight line cannot fit this pattern effectively.

The goal of the neural network is to learn a function capable of representing this behavior.

ReLU Activation Function

ReLU stands for:

Rectified Linear Unit

Its definition is:

ReLU(x) = max(0,x)

This means:

For negative values:

ReLU(-5) = 0
ReLU(-2) = 0
ReLU(-1) = 0

For positive values:

ReLU(1) = 1
ReLU(2) = 2
ReLU(5) = 5

Negative inputs become zero.

Positive inputs remain unchanged.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,500)

relu = np.maximum(0,x)

plt.figure(figsize=(8,5))

plt.plot(x,relu)

plt.title("ReLU Activation Function")

plt.xlabel("Input")

plt.ylabel("Output")

plt.grid()

plt.show()
```

Observation

The graph contains two regions:

A flat horizontal section for negative inputs.
A straight line with slope 1 for positive inputs.

Unlike Softplus, ReLU is not curved.

the first hidden neuron uses:

Weight = 1.70
Bias = -0.85

The hidden neuron computes:

x = (Dosage × 1.70) - 0.85

For dosage = 0:

x = (0 × 1.70) - 0.85

x = -0.85

Applying ReLU:

ReLU(-0.85)

= 0

Output becomes zero.

For dosage = 0.2:

x = (0.2 × 1.70) - 0.85

x = -0.51

Applying ReLU:

ReLU(-0.51)

= 0

Again, output remains zero.

For dosage = 0.6:

x = (0.6 × 1.70) - 0.85

x = 0.17

Applying ReLU:

ReLU(0.17)

= 0.17

The neuron now becomes active.



```python
dosage = np.linspace(0,1,100)

x_hidden = dosage*1.70 - 0.85

y_hidden = np.maximum(0,x_hidden)

plt.figure(figsize=(8,5))

plt.plot(dosage,y_hidden)

plt.title("Output of Hidden Node 1")

plt.xlabel("Dosage")

plt.ylabel("Activation")

plt.grid()

plt.show()
```

Observation

The hidden node remains inactive for small dosage values.

Once the transformed input becomes positive, the neuron begins producing output.

This creates a bend in the function.

Scaling the Hidden Node
Then we multiply the hidden node output by:

Weight = -40.8

This produces a new transformed curve.

Weights can:

Stretch functions
Compress functions
Flip functions
Change slopes

Second Hidden Node

A second hidden neuron processes the same dosage values.

However, it uses different weights and biases.

As a result:

It creates a different shape.
It extracts different information from the input.

Each neuron learns a unique feature.

Combining Hidden Nodes

The outputs from both hidden nodes are added together.

The neural network prediction becomes:

Output

=

Hidden Node 1 Contribution

+

Hidden Node 2 Contribution

+

Final Bias

This combination allows neural networks to create complex functions from simple components.

Why ReLU Works

Although ReLU is extremely simple, neural networks do not rely on a single ReLU.

Instead:

Multiple neurons are used.
Different weights are learned.
Different biases are learned.
Outputs are combined.

The resulting function can become highly complex.

This is why modern deep learning models work so effectively.

Advantages of ReLU
Very simple to compute.
Faster than sigmoid and tanh.
Helps reduce vanishing gradient problems.
Works well in deep networks.
Widely used in CNNs and modern architectures.
Limitations of ReLU
Neurons can become permanently inactive.
Negative inputs always produce zero.
Dead neurons may stop learning.

These issues motivated alternatives such as:

Leaky ReLU
ELU
GELU
