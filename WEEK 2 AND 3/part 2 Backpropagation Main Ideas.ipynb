Training a neural network means finding the optimal values of weights and biases that minimize prediction error.

Backpropagation is the algorithm used to calculate how much each weight and bias contributes to the total error. Once these contributions are known, gradient descent is used to update the parameters and improve predictions.

Backpropagation is the reason modern neural networks can learn from data.

Why Neural Networks Need Training

Initially, weights and biases are assigned random values.

Because of these random values, the predictions produced by the neural network are usually poor.

The goal of training is to gradually adjust these parameters until the network produces accurate predictions.

Forward Propagation

The first stage of neural network computation is forward propagation.

During forward propagation:

Input values are provided to the network.
Inputs are multiplied by weights.
Biases are added.
Activation functions are applied.
Predictions are generated.

These predictions are then compared with actual values.

Residuals

Residuals represent prediction errors.

Residual = Observed Value − Predicted Value

A residual close to zero indicates a good prediction.

Large residuals indicate poor predictions.




Sum of Squared Residuals (SSR)

To measure total error across all observations, residuals are squared and added together.

SSR = Σ(Observed − Predicted)²

Squaring ensures:

Positive and negative errors do not cancel each other.
Large errors receive larger penalties.

The objective of training is to minimize SSR.

Gradient Descent

Gradient descent is an optimization algorithm used to reduce error.

The basic idea is:

Calculate the slope of the loss function.
Determine the direction of steepest decrease.
Take a step in that direction.
Repeat until the error becomes small.




Learning Rate

The learning rate determines the size of each update step.

A very large learning rate may overshoot the minimum.

A very small learning rate may make training extremely slow.

Selecting an appropriate learning rate is critical for successful training.



Why Derivatives Are Needed

Derivatives tell us how sensitive the error is to changes in parameters.

If changing a weight significantly affects the error, the derivative will be large.

If changing a weight barely affects the error, the derivative will be small.

Backpropagation relies on derivatives to determine parameter updates



Chain Rule

Neural networks contain multiple layers of computation.

A weight influences the prediction indirectly through many intermediate calculations.

The chain rule allows us to calculate derivatives through these layers.

The chain rule is therefore the mathematical foundation of backpropagation.



Updating Weights

After calculating derivatives, weights are updated using:

New Weight = Old Weight − Learning Rate × Derivative

If the derivative is positive, the weight decreases.

If the derivative is negative, the weight increases.

The update always attempts to reduce error.


Updating Biases

Biases are updated in exactly the same way:

New Bias = Old Bias − Learning Rate × Derivative

Biases shift activation functions and therefore influence predictions.


Repeated Optimization

Training does not occur in a single step.

The process is repeated many times:

Forward propagation
Error calculation
Backpropagation
Parameter update

These repeated cycles gradually improve model performance.
