Unlike output-layer parameters, hidden-layer parameters influence predictions indirectly through activation functions.

This requires extensive use of the chain rule.

Hidden Layer Parameters

The parameters are:

W1
B1
W2
B2

These parameters belong to hidden neurons.

Why Hidden Layers Are Harder

Output-layer weights directly influence predictions.

Hidden-layer weights influence predictions indirectly through:

Input → Hidden Node → Activation Function → Output Layer → Prediction → Loss

Because of these intermediate steps, longer chain-rule calculations are required.

Softplus Activation Function

The hidden neurons use the Softplus activation function:

Softplus(x) = log(1 + eˣ)

The derivative becomes:

eˣ / (1 + eˣ)

This derivative appears repeatedly throughout the derivation.

Deriving dSSR/dB1

Bias B1 follows the same chain-rule path as W1.

The only difference is the final derivative term.

Because B1 is added directly:

dX1/dB1 = 1

Deriving dSSR/dW2

The derivation mirrors W1.

The second hidden neuron is used instead of the first.

The same chain-rule structure applies.

Deriving dSSR/dB2

The derivation mirrors B1.

Only variable names change.

Weight Initialization

Weights are initialized using a standard normal distribution.

This is only one of several possible initialization strategies.

Different initialization methods may affect training performance.

Bias Initialization

Biases are initialized to zero.

This is a common practice in neural network training.

Gradient Descent Optimization

After all derivatives are computed:

Gradients are calculated.
Step sizes are determined.
Parameters are updated.
Predictions are recalculated.

The process repeats until improvement becomes negligible.
