Introduction

Neural Networks are one of the most popular machine learning algorithms.
They are capable of learning complex patterns and non-linear relationships
between inputs and outputs.

Unlike linear regression, which can only fit straight lines,
neural networks can fit highly complex curves and decision boundaries.

In this notebook, we explore the basic structure of a neural network,
activation functions, hidden layers, weights, biases, and how neural
networks transform data to make predictions.

Simple Dataset

of a simple drug dosage dataset.

Low dosage -> ineffective
Medium dosage -> effective
High dosage -> ineffective

The goal is to predict effectiveness from dosage.

```python
import numpy as np
import matplotlib.pyplot as plt
dosage = np.array([0, 0.5, 1.0])

effectiveness = np.array([0, 1, 0])

plt.scatter(dosage, effectiveness)

plt.xlabel("Dosage")
plt.ylabel("Effectiveness")

plt.title("Drug Dosage Dataset")

plt.show()
```

Why Linear Models Fail

A straight line cannot accurately capture the pattern in this dataset.

The relationship is non-linear because effectiveness increases
and then decreases.

This motivates the use of neural networks.

```python
x = np.linspace(0,1,100)

y = -4*(x-0.5)**2 + 1

plt.plot(x,y)

plt.scatter(dosage,effectiveness)

plt.title("Example Non-Linear Relationship")

plt.show()
```

Basic Neural Network Structure

A neural network consists of:

1. Input Layer
2. Hidden Layer
3. Output Layer

Information flows through these layers using weights and biases.

Each hidden node applies an activation function.

Weights and Biases

Weights determine the importance of inputs.

Biases shift activation functions left or right.

Together, weights and biases control the shape learned by
the neural network.

Activation Functions

Activation functions introduce non-linearity.

Without activation functions, neural networks behave like
linear models regardless of depth.



Softplus Function

Softplus is defined as:

Softplus(x) = log(1 + e^x)

It is a smooth approximation of ReLU and was used in the
example neural network.


The first hidden neuron transforms dosage values into
x-coordinates for the activation function.

x = input × weight + bias

The transformed values are then passed through Softplus

Scaling Hidden Node Outputs

After applying Softplus, the outputs are multiplied
by another weight.

This stretches, flips, or shrinks the curve.

Second Hidden Node

The second hidden node uses different weights and biases.

As a result, it extracts a different portion of the
activation function.

Combining Hidden Nodes

The neural network adds the outputs of the hidden nodes.

This creates more complex shapes than either hidden node
could create individually.

Final Bias

The final bias shifts the entire curve upward or downward.

This allows the neural network to better fit the data.
