Why Softmax is Needed

Raw neural network outputs only indicate relative preference. They do not tell us how certain the model is about a prediction.

For example:

Output A:

Setosa = 10

Versicolor = 9

Virginica = 8

Output B:

Setosa = 10

Versicolor = 0

Virginica = -50

In both cases, Setosa has the highest score. However, the second prediction is clearly much more confident. Softmax converts these scores into probabilities and captures this difference in confidence.

Dependence Between Classes

A unique property of Softmax is that each probability depends on all output scores.

For example, the probability of Setosa depends not only on the Setosa score but also on the Versicolor and Virginica scores.

As a result:

Increasing one score affects all probabilities.
The probabilities compete with one another.
The sum of probabilities always remains equal to 1.

This makes the Softmax derivative more complex than many other activation functions.

Why Softmax Derivatives Are Needed

Neural networks learn using backpropagation and gradient descent.

Backpropagation requires derivatives that show how changes in parameters affect prediction error.

Since probabilities are produced by Softmax, the derivative of Softmax must be calculated before gradients can be propagated backward through the network.

Without these derivatives, the network cannot learn.

Derivative with Respect to the Same Class

Consider how the probability of a class changes when its own score changes.

After differentiating the Softmax equation and simplifying:

dPi/dzi = Pi * (1 - Pi)

This derivative describes how the probability of a class changes when its own score changes.

Example:

Suppose:

Pi = 0.69

Then:

dPi/dzi

= 0.69 * (1 - 0.69)

= 0.69 * 0.31

= 0.2139

≈ 0.21

This derivative becomes smaller as the probability approaches 0 or 1.

Derivative with Respect to a Different Class

Now consider how the probability of one class changes when the score of another class changes.

The derivative becomes:

dPi/dzj = -Pi * Pj

where i ≠ j

Example:

Pi = 0.69

Pj = 0.10

Then:

dPi/dzj

= -(0.69 * 0.10)

= -0.069

≈ -0.07

The negative sign indicates competition between classes.

If the probability of one class increases, the probabilities of the remaining classes must decrease because the total probability must remain equal to 1.

Jacobian Matrix

Since every Softmax output depends on every input score, a single derivative is not enough.

All derivatives are organized into a Jacobian Matrix.

For a three-class problem:

            z1       z2       z3

P1 dP1/dz1 dP1/dz2 dP1/dz3

P2 dP2/dz1 dP2/dz2 dP2/dz3

P3 dP3/dz1 dP3/dz2 dP3/dz3

The diagonal entries contain:

Pi * (1 - Pi)

The off-diagonal entries contain:

-Pi * Pj

The Jacobian Matrix allows gradients to flow correctly through the Softmax layer during backpropagation.

Introduction to Cross Entropy

After obtaining probabilities from Softmax, a loss function is required to measure prediction error.

Cross Entropy is one of the most widely used loss functions in machine learning and deep learning.

It measures how different the predicted probability distribution is from the actual distribution.

Smaller values indicate better predictions.

Larger values indicate poorer predictions.

Why Not Use Squared Error?

Squared Error works well for regression problems but is less effective for classification.

Cross Entropy is preferred because:

It penalizes incorrect predictions more strongly.
It produces larger gradients when predictions are poor.
It helps the network learn faster.

As a result, most classification networks use Cross Entropy instead of Squared Error.

Cross Entropy Formula

For a single training example:

Loss = -log(Pcorrect)

where:

Pcorrect = predicted probability assigned to the correct class.

The better the prediction, the smaller the loss.

The worse the prediction, the larger the loss.

General Cross Entropy Formula

For multiple classes:

Loss = -Σ yi * log(Pi)

where:

yi = actual probability
Pi = predicted probability

Since only one class is correct, most terms become zero and the equation simplifies significantly.

Example 1: Setosa

Suppose the true class is Setosa.

Predicted probabilities:

Setosa = 0.57

Versicolor = 0.21

Virginica = 0.22

Cross Entropy Loss:

Loss = -log(0.57)

= 0.56

This value represents the prediction error for that observation.

Example 2: Virginica

Suppose the true class is Virginica.

Predicted probability:

P(Virginica) = 0.58

Cross Entropy Loss:

Loss = -log(0.58)

= 0.54

Example 3: Versicolor

Suppose the true class is Versicolor.

Predicted probability:

P(Versicolor) = 0.52

Cross Entropy Loss:

Loss = -log(0.52)

= 0.65

Total Loss

The total error for the neural network is obtained by summing the individual Cross Entropy values.

Example:

Total Loss

= 0.56 + 0.54 + 0.65

= 1.75

The goal of training is to minimize this total loss by adjusting the network parameters.

Why Cross Entropy Works Well

Cross Entropy strongly penalizes bad predictions.

When the predicted probability of the correct class approaches zero:

Loss becomes very large.
Gradients become larger.
Weight updates become larger.

This allows the neural network to correct mistakes more aggressively and learn faster.

Comparison with Squared Error

For poor predictions:

Squared Error increases gradually.

Cross Entropy increases much more rapidly.

Because gradient descent depends on derivatives, the larger gradients produced by Cross Entropy help the network move toward better solutions more efficiently.

This is one of the main reasons Cross Entropy is preferred for classification tasks.

Softmax and Cross Entropy Together

Modern classification networks typically follow the sequence:

Raw Scores

↓

Softmax

↓

Probabilities

↓

Cross Entropy

↓

Loss

↓

Backpropagation

↓

Weight Updates

Softmax converts scores into probabilities.

Cross Entropy measures prediction error.

Backpropagation uses derivatives to update parameters.

Together, these components allow neural networks to learn effectively from data.
