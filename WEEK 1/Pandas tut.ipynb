Filtering Data

```python
# Show Pokemon with HP greater than 80

df[df["HP"] > 80]

# Show Pokemon with Attack greater than 100

df[df["Attack"] > 100]

df[df["Type 1"] == "Fire"]

# Water or Grass type Pokemon

df[
    (df["Type 1"] == "Water")
    |
    (df["Type 1"] == "Grass")
]

# Check whether Type 1 belongs
# to the given list

df[
    df["Type 1"].isin(
        ["Fire","Water","Grass"]
    )
]

df[df["Attack"] > 100][
    ["Name","Attack"]
]
```

Adding and Removing Columns

```python
df["Total"] = (
    df["HP"]
    + df["Attack"]
    + df["Defense"]
)
```

```python
df["Attack_Percent"] = (
    df["Attack"] /
    df["Total"]
)*100
```

```python
df.drop(
    columns=["Attack_Percent"]
)
```

```python
df.rename(
    columns={
       "HP":"Health"
    }
)
```

Combination

```python
import pandas as pd

df = pd.read_csv(
   "C:\\Users\\DELL\\Desktop\\SOC 2026\\WEEK 1\\pokemon_data.csv"
)

df["Total"] = (
   df["HP"]
   + df["Attack"]
   + df["Defense"]
)

top10 = df.sort_values(
   "Total",
   ascending=False
).head(10)

print(top10[
   ["Name","Total"]
])
```

```python
top10.to_csv(
   "C:\\Users\\DELL\\Desktop\\SOC 2026\\WEEK 1\\pokemon_data.csv",
   index=False
)
```
