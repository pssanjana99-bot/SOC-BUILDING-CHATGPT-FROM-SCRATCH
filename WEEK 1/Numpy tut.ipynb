```python
import numpy as np

marks = np.array([85,90,78,92])

print(marks)
```

BASICS

```python
print(marks.ndim)
print(marks.shape)
print(marks.dtype)
print(marks.itemsize)
print(marks.nbytes)
print(marks.size)
```

Indexing and Slicing

```python
attendance = np.array([
    [92,95,88,90],
    [85,87,89,91]
])

attendance[0]
attendance[:,2]
attendance[1,3]
attendance[0,1:4:2]
```

3D Arrays

```python
lab_scores = np.array([
    [[12,15],[18,20]],
    [[25,30],[35,40]]
])

print(lab_scores)

lab_scores[0,1,1]

lab_scores[:,1,:]
```

Array Initialization

```python
np.zeros((3,4))

np.ones((2,3))

np.full((3,3),7)

np.identity(4)

np.random.rand(4,4)

np.random.randint(1,100,size=(5,5))

np.repeat([[1,2]],4,axis=0)
```

Pattern Problem

```python
pattern = np.ones((7,7))

inner = np.zeros((5,5))

inner[2,2] = 9

pattern[1:-1,1:-1] = inner

print(pattern)
```

Copying Arrays

```python
a = np.array([10,20,30])

b = a.copy()

b[0] = 100

print(a)
print(b)
```

Mathematics

```python
a = np.array([2,4,6,8])

a + 3

a - 1

a * 2

a / 2

a ** 2

np.sin(a)

np.cos(a)

np.tan(a)
```

Linear Algebra

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

np.matmul(A,B)

np.linalg.det(A)

np.trace(A)

np.linalg.inv(A)
```

Statistics

```python
scores = np.array([
    [80,90,85],
    [75,95,88]
])

np.min(scores)

np.max(scores)

np.mean(scores)

np.sum(scores)

np.std(scores)
```

Reshape

```python
arr = np.arange(12)

arr.reshape(3,4)

arr.reshape(2,6)

arr.reshape(4,3)
```

Stack

```python
a = np.array([1,2,3])

b = np.array([4,5,6])

np.vstack((a,b))

np.hstack((a.reshape(3,1),
           b.reshape(3,1)))
```
