```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Creating the First Graph

```python
# Basic x and y values

x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Plot a line graph

plt.plot(x,y)

# Display graph

plt.show()
```

Adding Title and Axis Labels

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)

# Add graph title

plt.title("Simple Line Graph")

# Add axis labels

plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

Changing Font Size and Style

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

font1 = {
    "family":"Arial",
    "size":16
}

plt.plot(x,y)

plt.title(
    "Customized Title",
    fontdict=font1
)

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

Custom Tick Marks

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)

# Customize x-axis ticks

plt.xticks([0,1,2,3,4])

# Customize y-axis ticks

plt.yticks([0,2,4,6,8])

plt.show()
```

Adding a Legend

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    label="y = 2x"
)

plt.legend()

plt.show()
```

Changing Color

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    color="red"
)

plt.show()
```

Changing Line Width

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    linewidth=4
)

plt.show()
```

Adding Markers

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    marker="o"
)

plt.show()
```

Marker Size

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    marker="o",
    markersize=10
)

plt.show()
```

Different Marker Styles

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    marker="^"
)

plt.show()
```

Line Styles

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    linestyle="--"
)

plt.show()
```

Shorthand Notation

```python
# r = red
# o = circle marker
# -- = dashed line

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(
    x,
    y,
    "ro--"
)

plt.show()
```

Plotting Multiple Lines

```python
x = np.arange(0,6)

y1 = x*2
y2 = x**2

plt.plot(
    x,
    y1,
    label="2x"
)

plt.plot(
    x,
    y2,
    label="x²"
)

plt.legend()

plt.show()
```

Using NumPy Arrays

```python
x = np.arange(
    0,
    5,
    0.5
)

y = x**2

plt.plot(x,y)

plt.show()
```

Solid and Dashed Portions


```python
x = np.arange(
    0,
    5,
    0.5
)

y = x**2

plt.plot(
    x[:6],
    y[:6],
    label="Observed"
)

plt.plot(
    x[5:],
    y[5:],
    "--",
    label="Projected"
)

plt.legend()

plt.show()
```

Resizing the Figure

```python
plt.figure(
    figsize=(8,5),
    dpi=100
)

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)

plt.show()
```

Grid

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)

plt.grid()

plt.show()
```

Saving Graphs

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)

plt.savefig(
    "study_graph.png",
    dpi=300
)

plt.show()
```

Bar Chart

```python
subjects = [
    "Math",
    "Physics",
    "Chemistry"
]

marks = [
    88,
    92,
    84
]

plt.bar(
    subjects,
    marks
)

plt.title(
    "Subject Marks"
)

plt.show()
```

Colored Bar Chart


```python
subjects = [
    "Math",
    "Physics",
    "Chemistry"
]

marks = [
    88,
    92,
    84
]

plt.bar(
    subjects,
    marks,
    color=[
        "red",
        "blue",
        "green"
    ]
)

plt.show()
```

Bar Chart with Patterns

```python
subjects = [
    "Math",
    "Physics",
    "Chemistry"
]

marks = [
    88,
    92,
    84
]

bars = plt.bar(
    subjects,
    marks
)

bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")

plt.show()
```

Histogram

```python
scores = [
    60,65,70,72,
    80,85,88,90,
    92,95
]

plt.hist(
    scores,
    bins=5
)

plt.title(
    "Score Distribution"
)

plt.show()
```
