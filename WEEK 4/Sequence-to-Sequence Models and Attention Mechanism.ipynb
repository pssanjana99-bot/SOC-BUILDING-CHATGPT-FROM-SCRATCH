Motivation Behind Seq2Seq Models

Consider translating a sentence from English to French. The input and output sequences may have different lengths and structures. Traditional neural networks require fixed-size inputs and outputs, making them unsuitable for such tasks. Seq2Seq models solve this problem by processing sequences one element at a time and generating outputs sequentially. This allows them to handle variable-length inputs and outputs naturally.

Encoder

The Encoder is responsible for reading the entire input sequence and converting it into a numerical representation. As each word is processed, information accumulates within the hidden states. By the end of the sequence, the Encoder has captured the overall meaning of the sentence. This final representation acts as a summary of the entire input and is passed to the Decoder.

Context Vector

The final hidden state generated by the Encoder is known as the Context Vector. It acts as a compressed representation of the entire input sequence. The Decoder relies on this vector to understand the meaning of the input and generate the appropriate output sequence. The Context Vector serves as the bridge between the Encoder and Decoder.

Decoder

The Decoder receives the Context Vector and generates the output sequence one word at a time. Each generated word depends on both the Context Vector and the words that have already been generated. The Decoder continues producing words until it generates an end-of-sequence token. Through this process, the model can create complete translations, summaries or responses.

Example: Machine Translation

In a machine translation task, the Encoder processes an English sentence and creates a Context Vector containing its meaning. The Decoder then uses this information to generate the corresponding sentence in another language. Each output word is generated sequentially, allowing the model to produce coherent translations.

Teacher Forcing

Training Seq2Seq models can be difficult because an incorrect prediction early in the sequence may affect all subsequent predictions. Teacher Forcing is a technique used during training to overcome this problem. Instead of feeding the model's predicted word back into the Decoder, the correct word from the training data is provided. This helps the model learn faster and improves training stability.

Limitation of Basic Seq2Seq Models

The biggest limitation of the original Seq2Seq architecture is the Context Vector bottleneck. Regardless of sentence length, all information from the input sequence must be compressed into a single vector. This works reasonably well for short sentences but becomes problematic for longer sequences. Important information may be lost during compression, causing the Decoder to generate lower-quality outputs. As sentence length increases, this limitation becomes more severe.

Motivation for Attention

The Attention Mechanism was introduced to solve the Context Vector bottleneck. Instead of forcing the Decoder to rely entirely on a single vector, Attention allows it to access all Encoder hidden states whenever needed. This means the Decoder can revisit different parts of the input sequence while generating each output word. The approach is similar to how humans read and translate sentences by repeatedly looking back at important words and phrases.

Core Idea of Attention

When generating an output word, the Decoder determines which parts of the input sequence are most relevant at that moment. Rather than treating all input words equally, the model assigns different levels of importance to different words. Words that are more relevant receive higher attention scores, while less relevant words receive lower scores. This allows the model to focus on the most useful information for the current prediction.

Attention Scores

Attention begins by calculating a score that measures how relevant each Encoder hidden state is to the current Decoder state. Higher scores indicate stronger relevance. These scores help the model identify which parts of the input sequence should receive the most focus during prediction.

Softmax Attention Weights

The raw attention scores are converted into probabilities using the Softmax function. These probabilities are known as attention weights. The weights indicate how much importance should be assigned to each input word. All attention weights add up to one, making them easy to interpret as relative importance values.

Dynamic Context Vector

Unlike the original Seq2Seq model, Attention creates a new Context Vector for every output word. This Context Vector is formed by combining Encoder hidden states according to their attention weights. Words receiving higher attention contribute more information, while words receiving lower attention contribute less. As a result, the Decoder gains access to the most relevant information at every step.

Alignment Between Input and Output

One of the most useful properties of Attention is that it automatically learns alignments between input and output words. During translation, the model learns which input words correspond to which output words. This improves translation quality and makes the model more interpretable because we can visualize which parts of the input influenced each prediction.

Advantages of Attention

Attention removes the need to compress an entire sentence into a single vector. The Decoder can access information from any part of the input sequence whenever needed, making it much more effective for long sentences. Attention also improves interpretability because attention weights reveal which words influenced a prediction. Furthermore, it enables the model to capture long-range dependencies that are difficult for traditional Seq2Seq models to learn.

From Attention to Transformers

The success of Attention led researchers to develop Transformer architectures. Instead of relying on recurrent networks such as RNNs and LSTMs, Transformers use Attention as their primary mechanism for processing sequences. Modern models such as BERT, GPT, T5 and many other large language models are built upon Attention-based architectures. This makes Attention one of the most influential developments in modern deep learning.
