The Problem with One-Hot Encoding

Suppose we have the following words:

Cat
Dog
Tiger
Car

Using One-Hot Encoding:

Cat   = [1,0,0,0]

Dog   = [0,1,0,0]

Tiger = [0,0,1,0]

Car   = [0,0,0,1]

Although Cat and Dog are semantically similar, their vectors appear completely unrelated.

The distance between:

Cat and Dog

is identical to the distance between:

Cat and Car

This means One-Hot Encoding fails to capture word meaning and relationships.

What Are Word Embeddings?

Word Embeddings are dense vector representations of words.

Instead of representing a word using thousands of zeros and a single one, words are represented using smaller vectors containing meaningful numerical values.

For example:

Cat   = [0.82, 0.15, 0.67]

Dog   = [0.79, 0.18, 0.71]

Tiger = [0.91, 0.10, 0.80]

Car   = [0.12, 0.95, 0.22]

Notice that Cat, Dog and Tiger have similar values, while Car appears very different.

This allows neural networks to identify semantic similarities between words.

Why Word Embeddings Are Useful

Word Embeddings allow machine learning models to understand relationships between words.

Words that appear in similar contexts often receive similar vector representations.

For example:

King

Queen

Prince

Princess

would be placed near one another in the embedding space.

Similarly:

Apple

Banana

Orange

would form another cluster.

These relationships help neural networks perform better on language tasks.

Embedding Space

The collection of all word vectors forms an embedding space.

Each word occupies a position in this space.

Words with similar meanings tend to be located close together, while unrelated words appear farther apart.

The embedding space allows mathematical operations to reveal linguistic relationships.

One famous example is:

King - Man + Woman ≈ Queen

This demonstrates that embeddings capture meaningful semantic information rather than simply storing arbitrary numbers.

Learning Word Embeddings

Instead of manually assigning vector values, embeddings are learned automatically during training.

The neural network adjusts word vectors so that words appearing in similar contexts develop similar representations.

As training progresses, the embedding space gradually captures semantic and syntactic relationships between words.

Word2Vec

Word2Vec is one of the most influential techniques for generating word embeddings.

Developed by Google, Word2Vec trains a shallow neural network to learn meaningful vector representations of words.

Rather than directly learning language rules, Word2Vec learns word relationships by observing how words appear together in text.

The resulting embeddings can then be used for many natural language processing tasks.

Main Idea Behind Word2Vec

The central idea of Word2Vec is:

Words appearing in similar contexts tend to have similar meanings.

For example:

The cat sat on the mat.

The dog sat on the mat.

Since "cat" and "dog" appear in nearly identical contexts, Word2Vec learns similar vector representations for them.

Continuous Bag of Words (CBOW)

CBOW predicts a target word using surrounding context words.

Suppose the sentence is:

The cat sat on the mat

If the target word is:

sat

then the input might be:

The

cat

on

the

The model uses surrounding words to predict the missing word.

By repeatedly performing this task, the network learns meaningful embeddings.

Example of CBOW

Context Words:

The

cat

on

the

Target Word:

sat

The network receives the context words and attempts to predict the missing center word.

Training continues until predictions improve.

Skip-Gram Model

Skip-Gram works in the opposite direction.

Instead of predicting the center word from surrounding words, it predicts surrounding words from the center word.

Using the sentence:

The cat sat on the mat

Input:

sat

Outputs:

cat

on

The model learns embeddings by repeatedly predicting neighboring words.

CBOW vs Skip-Gram

CBOW generally trains faster because it uses multiple context words to predict a single target word.

Skip-Gram is often better at learning representations for rare words because each word generates multiple training examples.

Both approaches produce useful embeddings and are widely used in natural language processing.

Negative Sampling

A challenge with Word2Vec is that vocabulary sizes can become extremely large.

Predicting every possible word during training is computationally expensive.

Negative Sampling solves this problem by updating only a small subset of words during each training step.

Instead of comparing against the entire vocabulary, the model compares the correct word with a small number of randomly selected incorrect words.

This dramatically improves training efficiency.

Measuring Similarity Between Words

Once embeddings are learned, similarity between words can be measured using cosine similarity.

Cosine similarity measures the angle between two vectors.

Values close to:

1

indicate strong similarity.

Values close to:

0

indicate little similarity.

Values close to:

-1

indicate opposite relationships.
